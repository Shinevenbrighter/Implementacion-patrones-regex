##Introducción

"Las expresiones regulares, también conocidas como **regex**, son un conjunto de patrones utilizados para buscar, editar y manipular texto y datos. El término se originó en la utilidad UNIX ed. Las regex tienen varios usos, incluida la validación de la entrada del usuario, la búsqueda de archivos y cadenas, y la sustitución de texto dentro de una cadena." (¿Qué Es Regex Y Cómo Usarlo? | Lenovo México, 2021)

En esta actividad se desarrolla un tipo "formulario" que permite validar diferentes tipos de datos personales mediante expresiones regulares (regex). El objetivo principal es garantizar que la información ingresada cumpla con un formato correcto antes de ser aceptada, simulando así un sistema de captura de datos seguro y estructurado.





##Programa de validación

El formulario solicita de manera secuencial los siguientes datos: nombre, apellidos, edad, fecha de nacimiento, teléfono, dirección, correo electrónico y CURP. Cada dato se verifica automáticamente con un patrón de validación específico, asegurando que se cumplan reglas de sintaxis y consistencia, por ejemplo:
- Que los nombres y apellidos inicien con mayúscula.

- Que la edad esté dentro de un rango lógico (1 a 120 años).

- Que el teléfono y correo electrónico tengan un formato válido.

- Que el CURP cumpla con la estructura oficial y sea consistente con el nombre y la fecha de nacimiento de la persona.

```python
%%writefile formulario.py
import re

# -------------------------------
# Funciones validadoras
# -------------------------------
def checknombre(my_str):
    patron = r"^[A-ZÁÉÍÓÚÑ][a-záéíóúñ]+(\s[A-ZÁÉÍÓÚÑ][a-záéíóúñ]+)*$"
    res = bool(re.fullmatch(patron, my_str))
    print(f'{my_str} "{patron}" -> {"Coincide" if res else "No coincide"}')
    return res

def checkedad(my_str):
    patron = r"^(?:[1-9][0-9]?|1[01][0-9]|120)$"
    res = bool(re.fullmatch(patron, my_str))
    print(f'{my_str} "{patron}" -> {"Coincide" if res else "No coincide"}')
    return res

def checkphone(my_str):
    patron = r"^\+?\d{1,3}?\s?\d{7,15}$"
    res = bool(re.fullmatch(patron, my_str))
    print(f'{my_str} "{patron}" -> {"Coincide" if res else "No coincide"}')
    return res

def checkdireccion(my_str):
    patron = r"^[\w\s\.,#\-]+$"
    res = bool(re.fullmatch(patron, my_str))
    print(f'{my_str} "{patron}" -> {"Coincide" if res else "No coincide"}')
    return res

def checkemail(my_str):
    patron = r"^[\w\.-]+@[\w\.-]+\.\w{2,}$"
    res = bool(re.fullmatch(patron, my_str))
    print(f'{my_str} "{patron}" -> {"Coincide" if res else "No coincide"}')
    return res

def checkcurp(my_str):
    patron = r"^[A-Z]{4}\d{6}[HM][A-Z]{5}[A-Z0-9]\d$"
    res = bool(re.fullmatch(patron, my_str))
    print(f'{my_str} "{patron}" -> {"Coincide" if res else "No coincide"}')
    return res

# -------------------------------
# Generar CURP base esperado
# -------------------------------
def generar_curp(nombre, apellidos, fecha, sexo="H", estado="DF"):
    apellido_paterno, apellido_materno = apellidos.split()
    nombre = nombre.strip().upper()

    curp = (
        apellido_paterno[0].upper() +
        re.sub(r"[^AEIOU]", "", apellido_paterno[1:].upper())[0] +
        apellido_materno[0].upper() +
        nombre[0].upper()
    )

    yyyy, mm, dd = fecha.split("-")
    curp += yyyy[2:] + mm + dd
    curp += sexo
    curp += estado

    def primera_consonante(palabra):
        cons = re.sub(r"[AEIOU]", "", palabra.upper()[1:])
        return cons[0] if cons else "X"

    curp += (
        primera_consonante(apellido_paterno) +
        primera_consonante(apellido_materno) +
        primera_consonante(nombre)
    )

    curp += "0A1"
    return curp

# -------------------------------
# Función para ejecutar formulario
# -------------------------------
def ejecutar_formulario():
    print("=== Formulario de Registro ===\n")

    # Nombre
    while True:
        nombre = input("★ Nombre(s): ")
        if checknombre(nombre):
            print()
            break
        print("Nombre inválido, intenta de nuevo.\n")

    # Apellidos
    while True:
        apellidos = input("★ Apellidos (Paterno Materno): ")
        validos = True
        for x in apellidos.split():
            if not checknombre(x):
                validos = False
        if validos:
            print()
            break
        print("Apellidos inválidos, intenta de nuevo.\n")

    # Edad
    while True:
        edad = input("★ Edad: ")
        if checkedad(edad):
            print()
            break
        print("Edad inválida, intenta de nuevo.\n")

    # Fecha de nacimiento
    patron_fecha = r"^\d{4}-\d{2}-\d{2}$"
    while True:
        fecha = input("★ Fecha de nacimiento (YYYY-MM-DD): ")
        if re.fullmatch(patron_fecha, fecha):
            print(f'{fecha} "{patron_fecha}" -> Coincide\n')
            break
        print("Formato de fecha inválido (ejemplo: 2000-05-12), intenta de nuevo.\n")

    # Teléfono
    while True:
        tel = input("★ Teléfono (Ejemplo: +52 5561084772): ")
        if checkphone(tel):
            print()
            break
        print("Teléfono inválido, intenta de nuevo.\n")

    # Dirección
    while True:
        direccion = input("★ Dirección: ")
        if checkdireccion(direccion):
            print()
            break
        print("Dirección inválida, intenta de nuevo.\n")

    # Email
    while True:
        email = input("★ Email: ")
        if checkemail(email):
            print()
            break
        print("Email inválido, intenta de nuevo.\n")

    # CURP
    curp_esperado = generar_curp(nombre, apellidos, fecha, sexo="M")
    print(f"(Referencia CURP esperada según tus datos: {curp_esperado})\n")

    while True:
        curp = input("★ CURP: ").upper()

        if not checkcurp(curp):
            print("CURP inválido, intenta de nuevo.\n")
            continue

        if curp.startswith(curp_esperado[:10]):
            break
        else:
            print("CURP válido pero NO coincide con nombre y fecha, intenta de nuevo.\n")

    print("\n === Todos los patrones pasaron, el registro ha sido completado ===\n")


# -------------------------------
# PROGRAMA PRINCIPAL
# -------------------------------
for i in range(1, 5):
    print(f"\n>>> Caso de prueba {i} <<<")
    ejecutar_formulario()
```
